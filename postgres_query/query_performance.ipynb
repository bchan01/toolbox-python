```python
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
```

```python
# Load Summary Dataset
df = pd.read_csv("out.csv")
df.head(5)
```

```python
style.use('ggplot')
ax = plt.gca()
df.plot(kind='line',x='Trial',y='ElapsedTime', color='green', ax=ax, figsize=(20,10))
plt.title('Query Performance (milliseconds)')
plt.xlabel('Trial')
plt.ylabel('Time (msec)')
plt.show()
```
